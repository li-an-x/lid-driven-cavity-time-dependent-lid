# LDC with time-dependent boundary conditions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, num = 1000)
x1 = np.linspace(1, 1.5, num = 1000)
y = (np.sin(2 * x * np.pi) - np.cos(x1 * np.pi)) * np.sin(3 * (x + x1) * np.pi)
# y = np.sin (2 * x * np.pi) 
plt.title("time-depenednt lid velocity")


plt.plot(y)
```
